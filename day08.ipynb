```python
from collections import Counter
import heapq
import math
```

```python
with open("input/day8.txt") as file:
    light_boxes = [[int(x) for x in line.split(",")] for line in file.read().split("\n")]
```

```python
queue = []
parent_list = [i for i in range(len(light_boxes))]

def parent(v):
    return v if parent_list[v] == v else parent(parent_list[v])

for i in range(len(light_boxes)):
    for j in range(i,len(light_boxes)):
        if i == j:
            continue
        heapq.heappush(queue, [sum((light_boxes[i][k] - light_boxes[j][k])**2 for k in range(3)), i, j])
```

#### Part 1

```python
queue_copy = queue.copy()
parent_list = [i for i in range(len(light_boxes))]

for _ in range(1000):
    _, u, v = heapq.heappop(queue_copy)
    if (pu := parent(u)) != (pv := parent(v)):
        parent_list[pv] = pu

math.prod(x[1] for x in Counter(parent(i) for i in range(len(light_boxes))).most_common(3))
```

#### Part 2

```python
queue_copy = queue.copy()
parent_list = [i for i in range(len(light_boxes))]

while len(set(parent(i) for i in range(len(light_boxes)))) != 1:
    _, u, v = heapq.heappop(queue_copy)
    if (pu := parent(u)) != (pv := parent(v)):
        parent_list[pv] = pu

light_boxes[u][0] * light_boxes[v][0]
```
