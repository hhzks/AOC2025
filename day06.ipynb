```python
import math
```

```python
problems_list = []
with open("input/day6.txt") as file:
    file_read = file.read().split("\n")
    spacing_list = []
    op_list = [file_read[-1][0]]
    index = 0
    space = 1
    for char in file_read[-1][1:]:
        if char in ("+", "*"):
            spacing_list.append(space)
            if char == "+":
                op_list.append("+")
            else:
                op_list.append("*")
            space = 0
        space += 1
        index += 1
    spacing_list.append(space+1)
    problems_list = [[] for _ in range(len(spacing_list))]

    for line in file_read[:-1]:
        curr_index = 0
        for i, curr_space in enumerate(spacing_list):
            problems_list[i].append(line[curr_index:curr_index+curr_space-1])
            curr_index += curr_space
```

#### Part 1:

```python
sum(sum(int(n) for n in problems) if op_list[i] == "+" else math.prod(int(n) for n in problems) for i, problems in enumerate(problems_list))
```

#### Part 2:

```python
ans = 0
for i,arr in enumerate(problems_list):
    working_ans = 0 if op_list[i] == "+" else 1

    for j in range(spacing_list[i]-1):
        current_num = ""
        for num in arr:
            current_num += num[j]
        working_ans = working_ans + int(current_num) if op_list[i] == "+" else working_ans * int(current_num)
    ans += working_ans

ans
```
