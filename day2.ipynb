```python
with open("input/day2.txt") as file:
    ranges = [[int(x) for x in line.split("-")] for line in file.read().split(",")]
```

#### Part 1:

```python
ans = 0
for start, end in ranges:
    for i in range(start, end+1):
        str_i = str(i)
        ans += i * (len(str_i) % 2 == 0 and str_i[:len(str_i)//2] == str_i[len(str_i)//2:])
ans
                
```

#### Part 2:

```python
ans = 0
for start, end in ranges:
    for i in range(start, end+1):
        str_i = str(i)
        is_invalid = False
        for j in range(1,1+(len(str_i)//2)):
            if len(str_i) % j != 0:
                continue
            initial_pattern = str_i[:j]
            mismatch_found = False
            for k in range(j,len(str_i),j):
                if str_i[k:k+j] != initial_pattern:
                    mismatch_found = True
                    break
            if not mismatch_found:
                is_invalid = True
                break
        ans += i * is_invalid 
ans
```
