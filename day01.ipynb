```python
with open("input/day1.txt") as file:
    rotations = [int(line[1:]) if line[0] == "R" else -int(line[1:]) for line in file.read().split("\n")]
```

#### Part 1:

```python
pos = 50
ans = 0
for rotation in rotations:
    pos = (rotation+pos) % 100
    ans += (pos == 0)
ans
```

#### Part 2:

```python
pos = 50
ans = 0
for rotation in rotations:
    ans -= (rotation < 0 and pos == 0)
    div, pos = divmod(rotation+pos, 100)
    ans += (rotation < 0 and pos == 0)
    ans += abs(div)
ans
```
