```python
from functools import cache
```

```python
with open("input/day11.txt") as file:
        adjacency_list = {line.replace(':', '').split()[0]: line.replace(':', '').split()[1:] for line in file.read().split("\n")}
```

#### Part 1

```python
@cache
def part_1(node):
    return 1 if node == "out" else sum(part_1(connected_node) for connected_node in adjacency_list[node])

part_1("you")
```

#### Part 2:

```python
@cache
def part_2(node, dac_seen = False, fft_seen = False):
    return int(dac_seen and fft_seen) if node == "out" else sum(part_2(connected_node, dac_seen or node == "dac", fft_seen or node == "fft") for connected_node in adjacency_list[node])

part_2("svr")
```
