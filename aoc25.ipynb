### Day 1: Mod safe

#### Input processing:

```python
with open("input/day1.txt") as file:
    rotations = [int(line[1:]) if line[0] == "R" else -int(line[1:]) for line in file.read().split("\n")]
```

#### Part 1:

```python
dial = 50
ans = 0
for rotation in rotations:
    dial = (rotation+dial) % 100
    ans += (dial == 0)
print(ans)
```

#### Part 2:

```python
dial = 50
ans = 0
for rotation in rotations:
    ans -= (rotation < 0 and dial == 0)
    div, dial = divmod(rotation+dial, 100)
    ans += (rotation < 0 and dial == 0)
    ans += abs(div)
print(ans)
```

### Day 2: ID Validity

#### Input Processing:

```python
with open("input/day2.txt") as file:
    ranges = [[int(x) for x in line.split("-")] for line in file.read().split(",")]
```

#### Part 1:

```python
ans = 0
for start, end in ranges:
    for i in range(start, end+1):
        str_i = str(i)
        ans += i * (len(str_i) % 2 == 0 and str_i[:len(str_i)//2] == str_i[len(str_i)//2:])
print(ans)
                
```

#### Part 2:

```python
ans = 0
for start, end in ranges:
    for i in range(start, end+1):
        str_i = str(i)
        is_invalid = False
        for j in range(1,1+(len(str_i)//2)):
            if len(str_i) % j != 0:
                continue
            initial_pattern = str_i[:j]
            flag = False
            for k in range(j,len(str_i),j):
                if str_i[k:k+j] != initial_pattern:
                    flag = True
                    break
            if not flag:
                is_invalid = True
                break
        ans += i * is_invalid 


print(ans)
```
