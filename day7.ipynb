```python
from functools import cache
```

```python
with open("input/day7.txt") as file:
    grid = [[char for char in line] for line in file.read().split("\n")]
```

#### Part 1:

```python
ans = 0
for i in range(len(grid)-1):
    for j in range(len(grid[0])):
        if grid[i][j] in ("S", "|") :
            if grid[i+1][j] == "^":
                grid[i+1][j-1] = "|"
                grid[i+1][j+1] = "|"
                ans += 1
            else:
                grid[i+1][j] = "|"
ans
```

#### Part 2:

```python

ans = 0
@cache
def dfs(starting_row, starting_column):
    if starting_row == len(grid):
        return 1
    grid_copy = [[x for x in grid[i]] for i in range(starting_row, len(grid))]
    grid_copy[0][starting_column] = "S"
    for i in range(len(grid_copy)-1):
        for j in range(len(grid[0])):
            if grid_copy[i][j] in ("S", "|") :
                if grid_copy[i+1][j] == "^":
                    return dfs(starting_row+i+1, j-1) + dfs(starting_row+i+1, j+1)
                else:
                    grid_copy[i+1][j] = "|"
    return 1

dfs(0,7)
```
