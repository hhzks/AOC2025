```python
with open("input/day3.txt") as file:
    banks = file.read().split("\n")
```

#### Part 1:

```python
ans = 0
for bank in banks:
    digit_one = max(bank[:len(bank)-1])
    digit_two = max(bank[bank.index(digit_one)+1:])
    ans += int(digit_one + digit_two)
ans
```

#### Part 2:

```python
def max_joltage(curr_digits,possible_choices):
    if len(curr_digits) == 12:
        return curr_digits
    if len(possible_choices) < 12 - len(curr_digits):
        return None
    
    sorted_choices = sorted(possible_choices, reverse=True)
    for choice in sorted_choices:
        ans = max_joltage(curr_digits+choice, possible_choices[possible_choices.index(choice)+1:])
        if ans:
            return ans
    return None
    

ans = 0
for bank in banks:
    ans += int(max_joltage("", bank))
print(ans)
```
