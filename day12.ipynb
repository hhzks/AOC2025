```python
with open('input/day12.txt') as file:
    *shape_sections, region_section = file.read().split('\n\n')
    shapes = [shape.split('\n') for shape in shape_sections]
    regions = []
    for line in region_section.strip().split('\n'):
        dimensions, *shape_counts = line.replace(":", "").split(" ")
        dimensions = dimensions.split("x")
        regions.append([[int(x) for x in dimensions], [int(x) for x in shape_counts]])
```

#### Part 1

```python
sum(sum(sum(regions[i][1][j] * line.count("#") for line in shapes[j]) for j in range(len(shapes))) <= regions[i][0][0] * regions[i][0][1] for i in range(len(regions)))
```
