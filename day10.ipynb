```python
import heapq
from scipy.optimize import linprog
```

```python
patterns = []
counters = []
buttons = []
with open("input/day10.txt") as file:
    for line in file.read().split("\n"):
        line = line.split(" ")
        patterns.append(line[0][1:-1])
        counters.append([int(x) for x in line[-1][1:-1].split(",")])
        buttons.append([[int(y) for y in x[1:-1].split(",")] for x in line[1:-1]])
```

#### Part 1:

```python
ans = 0
for i in range(len(patterns)):
    target = patterns[i]
    queue = [(0,'.' * len(target))]
    visited = {'.' * len(target)}
    min_presses = float('inf')
    
    while queue:
        presses, current_pattern = heapq.heappop(queue)

        if presses >= min_presses:
            continue
        
        if current_pattern == target:
            min_presses = min(min_presses, presses)
            continue
        
        for button in buttons[i]:
            new_pattern = list(current_pattern)
            for pos in button:
                new_pattern[pos] = '#' if new_pattern[pos] == '.' else '.'
            new_pattern = ''.join(new_pattern)
            
            if new_pattern not in visited:
                visited.add(new_pattern)
                heapq.heappush(queue,(presses + 1, new_pattern))
    
    ans += min_presses
ans
```

#### Part 2:

```python
sum(int(linprog([1] * len(buttons[i]), A_eq=[[1 if r in buttons[i][c] else 0 for c in range(len(buttons[i]))] for r in range(len(counters[i]))], b_eq=counters[i], bounds=[(0, None) for _ in range(len(buttons[i]))], method='highs', integrality=1).fun) for i in range(len(patterns)))
```
