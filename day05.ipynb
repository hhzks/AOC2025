```python
with open("input/day5.txt") as file:
    interval_string, ingredient_string = file.read().split("\n\n")
    intervals = [[int(x) for x in line.split("-")] for line in interval_string.split("\n")]
    ingredients = [int(x) for x in ingredient_string.split("\n")]
```

```python
intervals.sort()
merged_intervals = []
for interval in intervals:
    if not merged_intervals or merged_intervals[-1][1] < interval[0]:
        merged_intervals.append(interval)
    else:
        merged_intervals[-1][1] = max(interval[1], merged_intervals[-1][1])
```

#### Part 1:

```python
sum(any(ingredient in range(start, end+1) for start, end in merged_intervals) for ingredient in ingredients)
```

#### Part 2:

```python
sum(1 + end - start for start, end in merged_intervals)
```
