```python
from functools import cache
```

```python
with open("input/day7.txt") as file:
    grid = [[char for char in line] for line in file.read().split("\n")]
```

#### Part 1:

```python
ans = 0
beam_map = [False] * len(grid[0])
beam_map[grid[0].index("S")] = True
for i in range(1,len(grid)-1):
    for j in range(len(grid[0])):
        if beam_map[j] and grid[i+1][j] == "^":
                beam_map[j-1] = True
                beam_map[j+1] = True
                beam_map[j] = False
                ans += 1

ans
```

#### Part 2:

##### Version 1 (Tabulation)

```python
beam_map = [0] * len(grid[0])
beam_map[grid[0].index("S")] = 1
for i in range(1,len(grid)-1):
    for j in range(len(grid[0])):
        if beam_map[j] and grid[i+1][j] == "^":
                beam_map[j-1] += beam_map[j]
                beam_map[j+1] += beam_map[j]
                beam_map[j] = 0

sum(beam_map)
```

##### Vesion 2 (Memoization)

```python
@cache
def dfs(i, j):
    return 1 if i == len(grid) - 1 else dfs(i+1, j-1) + dfs(i-1, j+1) if grid[i+1][j] == "^" else dfs(i+1,j)

dfs(0, grid[0].index("S"))
```
