```python
with open("input/day9.txt") as file:
    red_tiles = [tuple(int(x) for x in line.split(",")) for line in file.read().split("\n")]
```

```python
max((1+a[0]-b[0]) * (1+a[1]-b[1]) for b in red_tiles for a in red_tiles)
```

```python

row_ranges = [[float("inf"), -float("inf")] for _ in range(max(y for _, y in red_tiles)+1)]
vertical_breaks = [[[]]]

for i in range(-1,len(red_tiles)-1):
    if red_tiles[i][0] == red_tiles[i+1][0]:
        for y in range(min(red_tiles[i][1], red_tiles[i+1][1]), max(red_tiles[i][1], red_tiles[i+1][1]) + 1):  
            row_ranges[y][0] = min(row_ranges[y][0], red_tiles[i][0])
            row_ranges[y][1] = max(row_ranges[y][1], red_tiles[i][0])
    
    elif red_tiles[i][1] == red_tiles[i+1][1]:
        row_ranges[red_tiles[i][1]][0] = min(row_ranges[red_tiles[i][1]][0], red_tiles[i][0], red_tiles[i+1][0])
        row_ranges[red_tiles[i][1]][1] = max(row_ranges[red_tiles[i][1]][1], red_tiles[i][0], red_tiles[i+1][0])

ans = 0
for i in range(len(red_tiles)):
    for j in range(i + 1, len(red_tiles)):
        if not any(min(red_tiles[i][0], red_tiles[j][0]) < row_ranges[y][0] or max(red_tiles[i][0], red_tiles[j][0]) > row_ranges[y][1] for y in range(min(red_tiles[i][1], red_tiles[j][1]), max(red_tiles[i][1], red_tiles[j][1]) + 1)):
            ans = max(ans,(1 + abs(red_tiles[i][0] - red_tiles[j][0])) * (1 + abs(red_tiles[i][1] - red_tiles[j][1])))

ans
```
